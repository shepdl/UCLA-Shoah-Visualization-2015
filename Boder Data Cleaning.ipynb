### Clean data

```python
del Counts["id"]
```

```python
Counts.head()
```

```python
Facts.columns
```

```python
boder_meta = pd.DataFrame()
boder_meta["interviewee_id_inner"] = Original_names["id"]
boder_meta["interviewee_last"] = Original_names["last_name"]
boder_meta["interviewee_first"] = Original_names["first_name"]
boder_meta_names = ["interviewee_id_outter", "interviewer_id", "diff", "ratio", "gender",
                    "language_interviewed", "age", 'religion', 'date_of_interview', 
                    'location_of_interview','nationality','last_alias1', 'first_alias1',
                    'last_alias2', 'first_alias2','interviewed_multi_times?', 
                    'where_want_to_go?','related_to','notes','additional_info']
for i in range(len(boder_meta_names)):
    boder_meta[boder_meta_names[i]] = [np.nan]*len(boder_meta)
```

```python
boder_meta.head()
```

```python
for i in range(len(boder_meta)):
    # match the Counts table
    for j in range(len(Counts)):
        if (boder_meta["interviewee_last"][i]==Counts["interviewee_last_name"][j] and
            boder_meta["interviewee_first"][i]==Counts["interviewee_first_name"][j]):
            boder_meta["interviewee_id_outter"][i]=Counts["interviewee"][j]
            boder_meta["interviewer_id"][i]=Counts["interviewer"][j]
            boder_meta["diff"][i]=Counts["diff"][j]
            boder_meta["ratio"][i]=Counts["ratio"][j]
    # match the Facts table
    for k in range(len(Facts)):
        if (boder_meta["interviewee_last"][i]==Facts["Last Name"][k] and
            boder_meta["interviewee_first"][i]==Facts["First Name"][k]):
            boder_meta["gender"][i]=Facts['Gender'][k]
            boder_meta["language_interviewed"][i]=Facts['Language Interviewed'][k]
            boder_meta["age"][i]=Facts['Age'][k]
            boder_meta["religion"][i]=Facts['Religion'][k]
            boder_meta['date_of_interview'][i]=Facts['Date of Interview'][k]
            boder_meta['location_of_interview'][i]=Facts['Location of Interview'][k]
            boder_meta['nationality'][i]=Facts['Nationality'][k]
            boder_meta['last_alias1'][i]=Facts['Last Alias1'][k]
            boder_meta['first_alias1'][i]=Facts['First Alias1'][k]
            boder_meta['last_alias2'][i]=Facts['Last Alias2'][k]
            boder_meta['first_alias2'][i]=Facts['First Alias2'][k]
            boder_meta['interviewed_multi_times?'][i]=Facts['Interviewed Multiple Times?'][k]
            boder_meta['where_want_to_go?'][i]=Facts['Where did they want to go?'][k]
            boder_meta['related_to'][i]=Facts['Related to?'][k]
            boder_meta['notes'][i]=Facts['Notes'][k]
            boder_meta['additional_info'][i]=Facts['Additional Information'][k]
```

### Export data and further cleaning

```python
boder_meta.to_csv("boder_meta.csv", index=False)
```
