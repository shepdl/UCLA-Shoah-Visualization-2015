```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('Boder_transcr_1_Ilmar_A.xml')
root = tree.getroot()
```

```python
root.tag
```

```python
root.attrib
```

```python
for line in root.iter('line'):
    print(line.attrib)
```

```python
# get all the timestamps under 'time'
time1=[]
for line in root.iter('line'):
    time1.append(float(line.get('time')))
```
