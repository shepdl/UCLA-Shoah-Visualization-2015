### Load packages and input data

```python
import pandas as pd
import numpy as np
import re
import docx
import os
```

```python
# reference: https://stackoverflow.com/questions/25228106/how-to-extract-text-from-an-existing-docx-file-using-python-docx
# import docx as a list of lines
def getText(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return fullText
```

- The auto converted docx files from OCR pdfs are not helpful. The main reason is that, there are two columns, header, and footers. 
- To get these data, I manually changed copied the data from the OCR pdfs and manually cleaned them into desired forms in docx.

```python
# Seperate \n into different elements
# return list of elements
# seperate by \n
# b is the list
def remove_line_char(b):
    index_b=[]
    i=0
    while i <len(b):
        if "\n" in b[i]:
            j=0
            while j < len(b[i])-1: 
                if b[i][j]=="\n":
                    index_b.append(b[i][:j])
                    b[i]=b[i][j+1:]
                    j=-1
                j+=1
            i-=1
        else:
            index_b.append(b[i])
        i+=1
    return index_b
```

```python
index1 = remove_line_char(index1)
index2 = remove_line_char(index2)
index3 = remove_line_char(index3)
```

```python
def to_range(str1):
    for i in range(len(str1)):
        if str1[i]=="-":
            a = int(str1[:i])
            b = int(str1[i+1:])
            return(list(range(a,b+1)))
```
