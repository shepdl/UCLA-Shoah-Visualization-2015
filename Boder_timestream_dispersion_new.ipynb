### Step 1: Get the data into df

```python
import pandas as pd
import numpy as np
```

```python
# import three updated indexes
df1 = pd.read_csv("Boder_Keyword_with_Parent - index1 (1).csv")
df2 = pd.read_csv("Boder_Keyword_with_Parent - index2 (1).csv")
df3 = pd.read_csv("Boder_Keyword_with_Parent - index3_v2.csv")
```

```python
# import meta and get the page range for each person
meta = pd.read_csv("Boder_meta_with_index.csv")
```

```python
# exclue nans num
df1 = df1[pd.notnull(df1['num'])].reset_index(drop=True)
df2 = df2[pd.notnull(df2['num'])].reset_index(drop=True)
df3 = df3[pd.notnull(df3['num'])].reset_index(drop=True)
```

```python
# change num to lists of intergers
import ast
df1.num = df1.num.apply(lambda x: ast.literal_eval(x.lstrip('u')))
df2.num = df2.num.apply(lambda x: ast.literal_eval(x.lstrip('u')))
df3.num = df3.num.apply(lambda x: ast.literal_eval(x.lstrip('u')))
```

```python
# we only want the testimonies with interviewee_index_range
meta = meta[pd.notnull(meta['interviewee_index_range'])].reset_index(drop=True)
```

```python
### Remake the df into lines, taking page numbers as "index"
meta.interviewee_index_list = meta.interviewee_index_list.apply(lambda x: ast.literal_eval(x.lstrip('u')))
df = pd.DataFrame()
id_list=[]
last_list=[]
first_list=[]
index_list=[]
```

```python
# add to lists
for i in range(len(meta)):
    id_list+=[meta.interviewee_id[i]]*len(meta.interviewee_index_list[i])
    last_list+=[meta.interviewee_last[i]]*len(meta.interviewee_index_list[i])
    first_list+=[meta.interviewee_first[i]]*len(meta.interviewee_index_list[i])
    index_list+=meta.interviewee_index_list[i]
```

```python
df["id"]=id_list
df["last_name"]=last_list
df["first_name"]=first_list
df["page_num"]=index_list
```

```python
# sort df in the order of page number
df = df.sort_values(by=["page_num"]).reset_index(drop=True)
```

```python
# add df1,2,3 together as df4
df4 = pd.concat([df1, df2, df3]).reset_index(drop=True)
```

```python
# expand the lists to df5
keyword_list=[]
num_list=[]
parent_list=[]
```

```python
for i in range(len(df4)):
    keyword_list+=[df4.keyword[i]]*len(df4.num[i])
    parent_list+=[df4.parent_keyword[i]]*len(df4.num[i])
    num_list+=list(df4.num[i])
```

```python
df5=pd.DataFrame()
df5["keyword"]=keyword_list
df5["num"]=num_list
df5["parent"]=parent_list
```

```python
# add the keywords and parent keywords
df["keyword"]=[np.nan]*len(df)
df["parent_keyword"]=[np.nan]*len(df)
```

```python
for i in range(len(df)):
    df.keyword[i]=list(df5[df5.num==df.page_num[i]].keyword)
    df.parent_keyword[i]=list(df5[df5.num==df.page_num[i]].parent)
```

```python
# helper function:
# input - inv_page_num, inv_total_pages, inv_seg_ratio
# output - inv_seg_num_list, inv_seg_val_list

def ratio_to_lists (page_num, total_pages, seg_ratio):
    
    ### starting points
    inv_sum = 100/total_pages # testimony_ratio = first inv_seg_ratio
    list1=[] # inv_seg_num_list
    list2=[] # inv_seg_val_list
    
    ### get list1
    if inv_sum.is_integer() or seg_ratio.is_integer():
        list1=list(range(int(seg_ratio-inv_sum)+1,int(seg_ratio)+1))
    else:
        list1=list(range(int(seg_ratio-inv_sum)+1,int(seg_ratio)+2))
    
    ### get list2
    for i in range(len(list1)):
    
        # get the segment number
        seg_num = list1[i]

        # calculate the last element
        if math.modf(seg_ratio)[0]-0<0.001:
            if inv_sum<1:
                last = round(inv_sum,3)
            else:
                last = 1.0
        else:
            last = round(seg_ratio-(seg_num-1),3)

        # dealing with different lengths
        if len(list1)==1:
            list2 = [last]
        elif len(list1)==2:
            first = round(inv_sum-last-1*(len(list1)-2),3)
            list2 = [first, last]
        else: # >=3
            first = inv_sum-last-1*(len(list1)-2)
            list2 = [first] + [1]*(len(list1)-2) + [last]
    
    ### return list1 and list2 in another list
    # access by ratio_to_lists (page_num, total_pages, seg_ratio) [0]
    # and ratio_to_lists (page_num, total_pages, seg_ratio) [1]
    return [list1,list2]
```

```python
df_tab["list1"]=[[]]*len(df_tab)
df_tab["list2"]=[[]]*len(df_tab)
```

```python
for i in range(len(df_tab)):
    inv_sum = df_tab.inv_page_num[i]
    total_pages = df_tab.inv_total_pages[i]
    seg_ratio = df_tab.inv_seg_ratio[i]
    df_tab.list1[i]=ratio_to_lists(inv_sum,total_pages,seg_ratio)[0]
    df_tab.list2[i]=ratio_to_lists(inv_sum,total_pages,seg_ratio)[1]
```
