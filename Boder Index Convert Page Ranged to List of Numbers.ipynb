In the column of interviewee_index_range, ranges are str while missing values are np.nan

```python
def to_range(str1):
    for i in range(len(str1)):
        if str1[i]=="-":
            a = int(str1[:i])
            b = int(str1[i+1:])
            return(list(range(a,b+1)))
```

```python
to_range(df.interviewee_index_range[2])
```

```python
df.interviewee_index_range = df.interviewee_index_range.fillna("")
```

```python
df
```

```python
df.interviewee_index_list = df.interviewee_index_range.apply(lambda x: to_range(x))
```

```python
df
```

```python
for i in range(len(df)):
    if df.interviewee_index_range[i]=="":
        df.interviewee_index_range[i]=np.nan
    if df.interviewee_index_list[i]==None:
        df.interviewee_index_list[i]=np.nan
```
