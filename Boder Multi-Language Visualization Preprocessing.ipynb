```python
# then change type of every element to int
df.german_lines = df.german_lines.apply(lambda y: [float(x) for x in y])
df.french_lines = df.french_lines.apply(lambda y: [float(x) for x in y])
df.english_lines = df.english_lines.apply(lambda y: [float(x) for x in y])
df.spanish_lines = df.spanish_lines.apply(lambda y: [float(x) for x in y])
df.polish_lines = df.polish_lines.apply(lambda y: [float(x) for x in y])
df.russian_lines = df.russian_lines.apply(lambda y: [float(x) for x in y])
df.lithuanian_lines = df.lithuanian_lines.apply(lambda y: [float(x) for x in y])
df.other_lines = df.other_lines.apply(lambda y: [float(x) for x in y])
```

```python
# get rid of duplicates 0.0 in every list of time_points column
df.time_points = df.time_points.apply(lambda x: list(set(x)))
```

We need to aggregate the data for visualization, basically into id, language, start, end, and length.

```python
# wrangle the data for Tableau
# start excludes the last point
df["start_points"]=df.time_points.apply(lambda x: x[:-1])
# end excludes the first point
df["end_points"]=df.time_points.apply(lambda x: x[1:])
```

```python
# length is end - start
time_length_list=[]
for i in range(len(df)):
    time_length_list.append([b_i - a_i for a_i, b_i in zip(df.start_points[i], df.end_points[i])])
df["time_length"]=time_length_list
```
